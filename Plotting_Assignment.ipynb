```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Set up the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Fuel Efficiency Data")
plt.show()
```

**Interpretation of the Data**

*   The engine size metrics, which are the cylinders and displacment strongly correlate with each other with the indication of dark red on the heat map. This means larger engines typically have more cylinders. They also have higher CO2 emissions


*   All efficiency measures (city, highway, and combined) form a positively correlated cluster that negatively correlates with engine size.


*   Environmental ratings (feScore and ghgScore) strongly correlate with each other and with efficiency metrics, while negatively correlating with engine size and emissions.


In general, larger engines consistently produce more emissions and lower efficiency across all driving conditions, while vehicles with better fuel economy inherently have better environmental ratings.


**When targeting ghgScore, these columns would be dropped:**
